```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Spamham = pd.read_csv('/content/drive/MyDrive/mail_data.csv')
```

```python
Spamham.head(15)
```

```python
X = Spamham['Message']
Spamham.loc[Spamham['Category'] == 'ham', 'Category',] = 0
Spamham.loc[Spamham['Category'] == 'spam', 'Category',] = 1
y = Spamham['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.15, random_state= 42)
```

```python
y_train = y_train.astype('int')
```

```python
y_test = y_test.astype('int')
```

```python
linearSVC = LinearSVC()
```

```python
logreg = LogisticRegression()
```

```python
feature_extraction = TfidfVectorizer(min_df= 1, stop_words= 'english', lowercase= True)
```

```python
X_train_feature_extraction = feature_extraction.fit_transform(X_train)
```

```python
X_test_feature_extraction = feature_extraction.transform(X_test)
```

```python
print(X_train_feature_extraction)
```

```python
print(X_test_feature_extraction)
```

```python
logreg.fit(X_train_feature_extraction, y_train)
```

**Training** **Accuracy**

```python
predict_training = logreg.predict(X_train_feature_extraction)
```

```python
training_accuracy = accuracy_score(y_train, predict_training)
```

```python
print("Tesing accuracy is ", training_accuracy)
```

**Testing Accuracy**

```python
predict_test = logreg.predict(X_test_feature_extraction)
```

```python
testing_accuracy = accuracy_score(y_test, predict_test)
```

```python
print("Tesing accuracy is ", testing_accuracy)
```
